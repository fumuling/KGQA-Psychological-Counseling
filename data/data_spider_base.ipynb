```python
import requests
import json
from lxml import etree
from selenium import webdriver
```

```python
url_list = ['http://jibing.wenyw.com/keshi-9.shtml',
            'http://jibing.wenyw.com/keshi-9/2.shtml',
            'http://jibing.wenyw.com/keshi-9/3.shtml',
            'http://jibing.wenyw.com/keshi-9/4.shtml']
options = webdriver.ChromeOptions()
options.add_argument('lang=zh_CN.UTF-8')
driver = webdriver.Chrome(chrome_options=options, executable_path='./chromedriver.exe')
```

```python
disease_name_list = []; disease_urlpostfix_list = []
for url in url_list:
    driver.get(url = url)
    page = driver.page_source
    page_html = etree.HTML(page)
    disease_name_list.extend(page_html.xpath('/html/body/div[3]/div[1]/ul[2]/div/ul[2]/li/a/text()'))
    disease_urlpostfix_list.extend(page_html.xpath('/html/body/div[3]/div[1]/ul[2]/div/ul[2]/li/a/@href'))
```

```python
dic = {}
for i in range(len(disease_name_list)):
    dic[disease_name_list[i]] = disease_urlpostfix_list[i]
```

```python
import json
url_prefix = 'http://jibing.wenyw.com'
```

```python
json_dic = {}
i = 1
for key, val in dic.items():
    disease_url = url_prefix + val
    driver.get(url = disease_url)
    page = driver.page_source
    page_html = etree.HTML(page)
    
    disease_info_dic = {}
    # 名称
    disease_info_dic['disease_name'] = key
    # 又名
    alternate_name = page_html.xpath('/html/body/div[3]/div[1]/div[1]/ul/ul[1]/div/ol')
    disease_info_dic['alternate_name'] = alternate_name[0].strip(['(',')','又名']) if len(symptom) != 0 else None
    # 病发部位
    pathogenic_site = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/div/dl/dd/a/text()')
    disease_info_dic['pathogenic_site'] = pathogenic_site[0]
    # 科室
    department = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[1]/dd/a/text()')
    if len(department) == 1:
        disease_info_dic['department'] = department[0]
    else:
        disease_info_dic['department'] = ','.join(department)
    # 症状
    symptom = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[2]/dd/text()')
    disease_info_dic['symptom'] = symptom[0] if len(symptom) != 0 else None
    # 检查项目
    check = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[3]/dd/text()')
    disease_info_dic['check'] = check[0] if len(check) != 0 else None
    # 并发症
    complication = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[4]/dd/text()')
    disease_info_dic['complication'] = complication[0] if len(complication) != 0 else None
    # 传染性
    infectivity = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[5]/dd/text()')
    disease_info_dic['infectivity'] = infectivity[0] if len(infectivity) != 0 else None
    # 遗传性
    heredity = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[6]/dd/text()')
    disease_info_dic['heredity'] = heredity[0] if len(heredity) != 0 else None
    # 易混淆病
    confusable_disease = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[7]/dd/text()')
    disease_info_dic['confusable_disease'] = confusable_disease[0] if len(confusable_disease) != 0 else None
    # 高发人群
    susceptible_crowd = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[1]/dl[8]/dd/text()')
    disease_info_dic['susceptible_crowd'] = susceptible_crowd[0] if len(susceptible_crowd) != 0 else None
    # 专家建议
    expert_remind = page_html.xpath('/html/body/div[3]/div[2]/ul[1]/div[2]/ul[2]/text()')
    disease_info_dic['expert_remind'] = expert_remind[0].strip() if len(expert_remind) != 0 else None
    # 词条数据写入字典
    json_dic[key] = disease_info_dic
    print('current index: %d item: %s'%(i, key))
    i += 1
```

```python
with open('data2.json', 'w', encoding='utf-8') as f:
    json.dump(json_dic, f, ensure_ascii=False)
```
